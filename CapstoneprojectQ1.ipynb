Import pandas and numpy

```python
import pandas as pd
import numpy as np
```

Import wiki page as csv data

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

print first lines of table

```python
df.head()
```

Remove all postcodes with Borough as 'Not assigned'

```python
NA = df[df['Borough'] == 'Not assigned'].index
df.drop(NA , inplace=True)
df
```

Check for any 'Neighbourhood''s Not assigned

```python
NAN = df[df['Neighbourhood'] == 'Not assigned'].index
NAN
```

None not assigned

Combine the Neighbourhoods with the same postcode and remove the duplicates

```python
df['Neighbourhood'] = df[['Postcode','Neighbourhood','Borough',]].groupby(['Borough', 'Postcode']).transform(lambda x: ','.join(x))
   
df.drop_duplicates(subset = 'Postcode', inplace =True)                      

df
```
